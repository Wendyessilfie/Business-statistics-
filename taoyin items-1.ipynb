```python
# Create a new DataFrame that includes all columns except 'ETIM', 'statistiek_hoofdgroep', and 'statistiek_subgroep'
TY_items_only = ty_no_duplicates.drop(columns=['ETIM', 'statistiek_hoofdgroep', 'statistiek_subgroep'])

# Replace all NaN values with 0
TY_items_only.fillna(0, inplace=True)
# Display the DataFrame to check changes
TY_items_only.head()
```

```python
# Exclude 'Articlenr' from scaling
TY_items_features = TY_items_only.drop(columns=['Articlenr'])

# Feature scaling
# Define the StandardScaler instance
scaler = StandardScaler()
# Fit the StandardScaler to the feature data
scaler.fit(TY_items_features)
# Scale the feature data and convert back to a DataFrame
TY_scaled = pd.DataFrame(scaler.transform(TY_items_features), columns=TY_items_features.columns)
# Display the result
TY_scaled.head()
```

```python
## Performing PCA
from sklearn.decomposition import PCA
pca = PCA()
pca_comps = pca.fit_transform(TY_scaled)                    # pca components
var_exp = pca.explained_variance_ratio_                     # variance explained ratio 
cum_var_exp = np.cumsum(var_exp)                            # cumulative variance
```
